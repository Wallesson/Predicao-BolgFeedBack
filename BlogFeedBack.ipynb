### Descrição do Dataset

#### Total de Atributos: 281
#### Caracteristicas dos Atributos: Inteiros e reais
#### Número de Instâncias de treino: 52.397

O Dataset é conhecido por BlogFeedback Data Set e está disponível em https://archive.ics.uci.edu/ml/datasets/BlogFeedback. Os atributos do 0 ao 270 são referentes a média, desvio padrão, mínimo, máximo, mediana, número de comentários em determinados intervalos de horas, etc. Já objetivo é apresentado na coluna 280. O objetivo descreve o número de comentários nas próximas 24 horas (em relação ao dataset).

O objetivo é prever quantos comentarios uma determinada publicação receberá. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('blogData_train.csv',names=range(0,281))
df_test = pd.read_csv('blogData_test.csv',names=range(0,281))
df
```

### Verificando dados

```python
print("Treino:",df.isna().sum())
print("Teste:",df_test.isna().sum())
```

### Separando em variaveis os valores de treino e teste

```python
X_train = df.loc[:, :'279']
y_train = df.loc[:,'280':]
X_test = df_test.loc[:, :'279']
y_test = df_test.loc[:,'280':]
```

### Normalizando os Dados 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
scaler.fit(X_test)
X_tr = scaler.transform(X_train)
X_tst = scaler.transform(X_test)
```

### Treinando o modelo com Regressão Linear

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_tr, y_train)
y_pred = reg.predict(X_tst)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("MSE", mean_squared_error(y_test, y_pred, squared=False))
print("R2", r2_score(y_test, y_pred))
```

```python
ref  = np.linspace(min(y_pred),max(y_pred),100)
plt.scatter(y_test,y_pred,s=10)
plt.plot(ref,ref,c='r')
plt.xlabel('Valor Predito')
plt.ylabel('Valor Real')
```

### Treinando o modelo com Ridge

```python
from sklearn.linear_model import Ridge
clf = Ridge(alpha=1.0)
clf.fit(X_tr, y_train)
y_pred = clf.predict(X_tst)
```

```python
score = clf.score(X_tr,y_train)
score
```

### Escolhendo um valor de lambda melhor

Como o valor de lambda tem que ser escolhido a mão, é criada uma lista com uma distribuição em um certo intervalo de valores para escolher o melhor.

```python
lambdas = [0.00001,0.0001,0.001,0.01,0.1,0.5,1,1.5,2,5]
scores = []
for i in range (len(lambdas)):
    clf = Ridge(alpha=i)
    clf.fit(X_tr,y_train)
    score = clf.score(X_tr,y_train)
    scores.append(score)
```

```python
plt.scatter(lambdas,scores)
plt.xlabel('Lambda')
plt.ylabel('Score')
```

```python
clf = Ridge(alpha=0.00001)
clf.fit(X_tr, y_train)
y_pred = clf.predict(X_tst)
```

```python
print("MSE", mean_squared_error(y_test, y_pred, squared=False))
print("R2", r2_score(y_test, y_pred))
```

```python
ref  = np.linspace(min(y_pred),max(y_pred),100)
plt.scatter(y_test,y_pred,s=10)
plt.plot(ref,ref,c='r')
plt.xlabel('Valor Predito')
plt.ylabel('Valor Real')
```

### Treinando o modelo com Redes Neurais

```python
from sklearn.neural_network import MLPRegressor
n_layers = 5#Número de camadas ocultas
n_elements_layers = 50 #Número de elementos ocultos em cada camada
layer_size = (n_elements_layers,)

for i in range(0,n_layers-1):
    layer_size+=(n_elements_layers,)
modelo = MLPRegressor(random_state=32, max_iter=100,hidden_layer_sizes=layer_size).fit(X_tr, y_train)
y_pred = modelo.predict(X_tst)
```

```python
print("MSE", mean_squared_error(y_test, y_pred, squared=False))
print("R2", r2_score(y_test, y_pred))
```

```python
ref  = np.linspace(min(y_pred),max(y_pred),100)
plt.scatter(y_test,y_pred,s=10)
plt.plot(ref,ref,c='r')
plt.xlabel('Valor Predito')
plt.ylabel('Valor Real')
```
